```python
import requests
from bs4 import BeautifulSoup
import zipfile
import pandas as pd
import os
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
url =  'https://dadosabertos.ans.gov.br/FTP/PDA/informacoes_consolidadas_de_beneficiarios-024/'
dir = 'Share_SP'
dir_adj = 'Share_ADJ_SP'
```

```python
#Extrac months
def extrair_links(url):
    reponse = requests.get(url)
    
    if reponse.status_code == 200:
        soup = BeautifulSoup(reponse.text, 'html.parser')
        
        links = soup.find_all('td') 

        links_ext = []
        for link in links:
            a_tag = link.find('a')
            if a_tag and 'href' in a_tag.attrs:
                links_ext.append(a_tag['href'])
        
        return links_ext
    else:

        print("Erro ao acessar a página:", reponse.status_code)
        return None
    
links = extrair_links(url)
```

```python
#here u put the month range desired
last_Year= links[-14:-1]
path_dir =  'Share_SP'
print(last_Year)
```

```python
import time
import random

def scrapping(urls, save_dir):
    os.makedirs(save_dir, exist_ok=True)

    for zip_url in urls:
        response = requests.get(zip_url)

        if response.status_code == 200:
            zip_filename = os.path.basename(zip_url)
            zip_path = os.path.join(save_dir, zip_filename)

            with open(zip_path, 'wb') as file:
                file.write(response.content)

            with zipfile.ZipFile(zip_path, 'r') as zip_ref:
                zip_ref.extractall(save_dir)

            os.remove(zip_path)
            
        else:
            print(f"Não foi possível baixar {zip_url}. Status code: {response.status_code}")

        time.sleep(random.uniform(1, 10))
```

```python
#Create URLs
for i in last_Year:
    response = requests.get(url + i) 

    if response.status_code == 200:   
        soup = BeautifulSoup(response.text, 'html.parser') 
        links = [a['href'] for td in soup.find_all('td') 
                          for a in td.find_all('a', href=True)]
        url_md = [url + i +link for link in links[1:]]
        scrap  = scrapping(url_md, dir)

    else:
        print(f"Erro ao acessar {url + i}: Status {response.status_code}")
```

```python
os.makedirs(dir_adj, exist_ok=True)

op = [327417, 335100, 350249, 366561]

for arquivo in os.listdir(dir):
    file_path = os.path.join(dir, arquivo)

    if arquivo.endswith('.csv'):
        try:
            df = pd.read_csv(file_path, sep=';', on_bad_lines='skip') 
            df_filtrado = df[df['CD_OPERADORA'].astype(int).isin(op)]

            if not df_filtrado.empty:
                caminho_destino = os.path.join(dir_adj, arquivo)
                df_filtrado.to_csv(caminho_destino )

            os.remove(file_path) 
            #print(f"Arquivo processado e removido: {arquivo}")

        except Exception as e:
            print(f"Erro ao processar {arquivo}: {e}")
```
